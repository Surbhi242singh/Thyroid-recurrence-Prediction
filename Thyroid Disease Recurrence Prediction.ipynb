## Train-test-split

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['recurred']),df['recurred'],test_size=0.2,random_state=2)
```

## Feature Engineering

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler

ct=ColumnTransformer(transformers=[          
    ('ct1',OneHotEncoder(drop='first',sparse=False),['gender','smoking','hx_smoking','hx_radiothreapy','focality']),
    ('ct2',OneHotEncoder(sparse=False),['thyroid_function','physical_examination','adenopathy','pathology','t','n','m','response']),
    ('ct3',OrdinalEncoder(categories=[['I','II','III','IVA','IVB']]),['stage']),
    ('ct4',OrdinalEncoder(categories=[['Low','Intermediate','High']]),['risk']),
    ('ct5',StandardScaler(),['age'])
],remainder='passthrough') 
```

```python
X_train=ct.fit_transform(X_train)
X_train
```

```python
X_test=ct.transform(X_test)
X_test
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
```

```python
y_train
```

```python
y_test
```

## Model fitting

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
```

```python
#instantiating classification models

models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Support Vector Machine': SVC(),
    'Gaussian Naive Bayes': GaussianNB(),
    'AdaBoost': AdaBoostClassifier(),
}
```

```python
#fitting models and generating classification reports

from sklearn.metrics import classification_report

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred)
    print(f'{name} Classification Report:\n{report}\n')
```
